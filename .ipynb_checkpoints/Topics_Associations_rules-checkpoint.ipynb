```python
import pandas as pd
import numpy as np 
```

```python
content = pd.read_csv('../data/projects_data.csv')
user_project = pd.read_csv('../server_data/user_projects_ip.csv')
```

```python
content
```

```python
len(content['project_topic_1'].unique())
```

```python
user_project = user_project[user_project.drop(['user','ip'],1).sum(axis=1)>=2]
```

```python
user_project
```

```python
topic_columns = list(set(list(content.project_topic_1.unique()) + list(content.project_topic_2.unique()) ))
topic_columns = [x for x in topic_columns if str(x) != 'nan']
user_topic = pd.DataFrame(columns=topic_columns,index = user_project.index )
user_topic
```

```python
for index ,user in user_project.iterrows():
    user_id = user['user']
    user_projects = user[user != 0].drop(['user','ip']).index.astype(int).values
    topics = content[content.project_id.isin(user_projects.astype(int))]
    user_topics_list = list(set(list(topics.project_topic_1.unique()) + list(topics.project_topic_2.unique()) ))
    user_topics_list = [x for x in user_topics_list if str(x) != 'nan']
    for topic in user_topics_list: 
        user_topic.loc[index][topic] = 1
user_topic.fillna(0, inplace = True)
user_topic[user_topic.sum(axis=1)>=2]
```

```python
user_topic
```

```python
from mlxtend.frequent_patterns import apriori, association_rules

frequent_itemsets = apriori(user_topic, min_support=0.01, use_colnames=True, max_len=2)
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=2)
```

```python
rules['antecedents'] = [list(a)[0] for a in rules['antecedents']]
rules['consequents'] = [list(a)[0] for a in rules['consequents']]
rules[['antecedents','consequents']].to_csv('../data/topics_associations_trained_2.csv',index=False)
```

```python
rules['consequents'] =  rules.consequents.apply(lambda x: list(x)[0]).astype(int)
rules['antecedents'] = rules.antecedents.apply(lambda x: list(x)[0]).astype(int)
rules = rules.iloc[::2]
rules=rules.sort_values(by=['support','confidence'],ascending=False)
rules = rules[['antecedents','consequents','support','confidence','lift']]
content =  content[['project_id','project_topic_1','project_topic_2']] 
rules = rules.merge(content,how='left',left_on='antecedents',right_on='project_id').drop('project_id',axis=1)
rules = rules.rename(columns={"project_topic_1": "antecedents_topic1", "project_topic_2": "antecedents_topic2"})
rules = rules.merge(content,how='left',left_on='consequents',right_on='project_id').drop('project_id',axis=1)
rules = rules.rename(columns={"project_topic_1": "consequents_topic1", "project_topic_2": "consequents_topic2"})
# rules.to_csv('Association_rules_data/Projects_association_with_topics.csv',index=False)
```

```python
import matplotlib

axarr = rules.hist(column=['support', 'confidence','lift'])
for ax in axarr.flatten():
    ax.set_ylabel("# one way pairs")
fig = matplotlib.pyplot.gcf()
fig.set_size_inches(18.5, 10.5)
fig.savefig('SciStarter_association_scores.png', dpi=100)
```
